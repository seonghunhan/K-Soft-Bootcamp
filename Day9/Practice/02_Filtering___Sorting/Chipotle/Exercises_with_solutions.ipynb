### Step 4. 가격(item_price)이 $10.00 이상인 데이터만 걸러내기

```python
# 새로운 변수에 item_price 열을 정리하고 float로 변환합니다
prices = chipo[['item_price']]

# chipo의 item_price를 위 prices 변수로 치환합니다
chipo.item_price = prices

# 'item_name', 'quantity', 'choice_description'의 중복을 제거하여 새로운 변수에 저장합니다.
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity', 'choice_description'])


# `chipo_filtered` 변수의 'quantity' 가 1인 데이터만 추려 새로운 변수에 저장합니다.
condition = (chipo_filtered.quantity == 1) & (chipo_filtered.quantity == True)

chipo_one_prod = chipo_filtered.loc[condition] # 필터링 된 row의 특정 컬럼만 추출


# 마지막으로 'chipo_one_prod'를 이용하여 item_price가 10달러 이상인 데이터를 걸러 봅니다 (위 과정은 중복을 피하기 위해 거친 단계입니다!)
# chipo_one_prod[???여기를 채워주세요???]
chipo_one_prod["item_price"] = chipo_one_prod["item_price"].str.replace(pat=r'$', repl=r'', regex=True)

a = chipo_one_prod.astype({'item_price' : float})

condition = (a.item_price >= float(10)) 

b = a.loc[condition]

print(b)

#아래 코드와 결과를 비교해보기
# chipo_one_prod[             ].item_name.nunique()
```
