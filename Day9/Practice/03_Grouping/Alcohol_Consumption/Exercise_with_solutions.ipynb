### Step 3. 'drinks' 변수에 다운로드 한 데이터셋 가져오기

```python
drinks = pd.read_csv('drinks.csv')
drinks.head()
```

### Step 4. 평균 이상의 `beer`을 마신 `continent` 찾아보기

```python
# mean_df = drinks['beer_servings'].mean()

# a = drinks.loc[drinks.beer_servings >= mean_df]
# b = a['continent']
# b

print(drinks.groupby('country').beer_servings.mean())
print(drinks.groupby('continent').beer_servings.mean())
```

### Step 5. 각 `continent`에 대해 와인 소비에 대한 통계를 print 하기

```python
import numpy as np

# a = drinks.groupby('continent')['wine_servings'].mean()
# b = drinks.groupby('continent')['wine_servings'].sum()

# a = drinks.groupby('continent').agg(np.mean)
# a
# # def my_mean(s):
# #     return np.mean(s)

# drinks.groupby('continent').agg({'mean': my_mean, 'quantity': np.sum})

drinks.groupby('continent').wine_servings.describe()
```

### Step 6. 각 `continent` 당 alcohol 소비에 대한 평균값을 column별로 도출해내기

```python
drinks.groupby('continent').mean()
```

### Step 7. 각 `continent` 당 alcohol 소비에 대한 중간값(median)을 column별로 도출해내기

```python
drinks.groupby('continent').median()
```

### Step 8. `spirit` 소비량의 최솟값, 최댓값, 평균값을 dataFrame 타입으로 불러오기

```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```
