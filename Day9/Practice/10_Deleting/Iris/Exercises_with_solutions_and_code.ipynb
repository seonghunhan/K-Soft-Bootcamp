### Step 3. 다운받은 dataset을 iris라는 변수에 할당하기

```python
raw_data = sns.load_dataset('iris')
df = raw_data.copy()
df
```

### Step 4. Dataset에 대한 열 생성하기

```python
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class

df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df
```

### Step 5.  Dataframe에 결측치가 있나요?

```python
df.isnull().sum()
```

### Step 6. `petal_length` 열의 10~29행 값을 `NaN`으로 설정하기

```python
df.iloc[10:30, 2] = np.nan
df.iloc[10:30, :]
```

### Step 7. `NaN` 값을 1.0으로 대체하기

```python
df = df.fillna(1.0)
df
```

### Step 8. `class` 열 삭제하기

```python
df = df.drop('class', axis=1)
df
```

### Step 9. 처음 3개 행을 `NaN` 으로 설정하기

```python
df.iloc[:3, :] = np.nan
df
```

### Step 10. `NaN` 값을 포함한 행 삭제하기

```python
df = df.dropna()
df
```

### Step 11. 인덱스가 0부터 시작하도록 재설정하기

```python
df.reset_index(inplace=True)
df
```
