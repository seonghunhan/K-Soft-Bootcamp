### Plotting Multiple Lines(여러 개의 선 Plot하기)

각 줄에 대해 `plt.plot`함수를 한 번씩 호출하여 동일한 그래프에 여러 줄을 표시할 수 있습니다. 간토의 사과와 오렌지의 수확량을 비교해 봅시다.

````python
years = range(2000, 2012) # 2000~2011 까지의 정수 범위 생성
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896 ]
# print(list(years))
# list(years)를 print하여 생성된 범위를 확인할 수 있음
````

````python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
````

### Chart Title and  Legend(그래프 제목과 범례)

여러 줄을 구별하기 위해 `plt.legend` 함수를 사용하여 범례를 그래프에 포함시킬 수 있습니다.  
`plt.title` 기능을 사용하여 그래프 제목을 설정할 수도 있습니다.

````python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']); # 순서도 중요함
````

### Line Markers

`plt.plot`의 `marker`인수를 사용하여 각 라인의 데이터 포인트에 대한 마커를 표시 할 수도 있습니다. Matplotlib는 원, 십자형, 사각형, 다이아몬드 등과 같은 다양한 마커를 제공합니다.  
우측 링크에서 마커 유형의 전체 목록을 찾을 수 있습니다.: https://matplotlib.org/3.1.1/api/markers_api.html .

````python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
````

### Styling Lines and Markers

`plt.plot`함수는 라인 및 마커 스타일을 지정하기 위한 인자들을 지원합니다.

- `color` 또는 `c`: 색깔이나 선을 정한다. ([supported colors](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
- `linestyle` 또는 `ls`: 점선이나 실선인지 선의 스타일을 정한다.
- `linewidth` 또는 `lw`: 선의 두께를 설정한다.
- `markersize` 또는 `ms`: 마커의 크기를 설정한다.
- `markeredgecolor` 또는 `mec`: 마커의 테두리 색을 정한다.
- `markeredgewidth` 또는 `mew`: 마커의 테두리 선의 두께를 정한다.
- `markerfacecolor` 또는 `mfc`: 마커의 색을 채운다.
- `alpha`: 플롯의 투명도를 조절한다.


 더 자세한 내용은 `plt.plot` Documentaion을 참고하길 바랍니다. : [https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) .


````python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
````

`fmt` 인자는 마커의 모양, 선 스타일 및 선 색상을 지정하는 약어를 제공합니다. `plt.plot`의 세 번째 인자로 전달하여 사용할 수 있습니다.

```
fmt = '[marker][line][color]'
```

````python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
````


만약 `fmt`에 선 스타일을 지정하지 않을 경우 , 마커만 그려지게 됩니다.

````python
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
````

## Scatter Plot(산점도)

산점도에서, 2개의 변수값은 2차원 좌표에 점으로 표시됩니다. 또한 세번째 변수를 사용하여 포인트의 크기나 색상을 결정할 수 있습니다. 

[Iris flower dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set): 세 종의 꽃에 대한 꽃받침과 꽃잎의 샘플 측정을 제공합니다.   
Iris 데이터셋은 Seaborn 라이브러리에 포함되어 있으며 Pandas 데이터 프레임으로 로드 할 수 있습니다.

````python
# Load data into a Pandas dataframe
flowers_df = sns.load_dataset("iris")
````

````python
flowers_df
````

````python
flowers_df.species
````

꽃받침 길이와 꽃받침 너비의 관계를 시각화 해 보겠습니다. 첫 번째 일은 `plt.plot`을 사용하여 선형 차트를 만드는 것 입니다.

````python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
````

데이터셋 내에 두 속성의 조합이 너무 많기 때문에 출력값은 그다지 유익한 정보를 가지고 있지 않습니다.  
위 그래프를 보면, 두 속성 사이에는 관계성이 없는 것 같습니다.  
산점도를 사용하여 `seaborn`모듈의 `scatterplot`함수를 사용하여 꽃받침 길이와 꽃받침 너비가 어떻게 달라지는지 시각화 할 수 있습니다.

````python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
````

### Adding Hues(색상 추가)

위 그림의 점이 일부 특이값이 있는 별개의 군집을 형성하는지 확인해봅시다. 꽃 종을 `hue`를 사용하여 점에 색깔을 줄 수 있습니다.  
`s` 인자를 사용하여 점을 더 크게 만들 수 있습니다.

````python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100); # hue 인자를 사용하여 종에 따른 색상을 추가함
````

색조를 추가하면 그래프에서 더 많은 정보를 얻을 수 있습니다. setosa의 꽃받침 길이는 더 짧지만, 꽃받침 너비는 더 높다는 것을 알 수 있습니다. 반대로 Virginica의 경우 그 반대의 모습을 보입니다.

### Customizing Seaborn Figures(Seaborn 그림 사용자 정의)

Seaborn은 내부적으로 Matplotlib의 그래프작성 함수를 사용하므로, `plt.figure` 및 `plt.title`과 같은 함수를 사용하여 Figure를 수정 할 수 있습니다.
<!--Since Seaborn uses Matplotlib's plotting functions internally, we can use functions like `plt.figure` and `plt.title` to modify the figure.-->

````python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species,
                s=100);
````

### Plotting using Pandas Data Frames(Pandas 데이터프레임을 이용한 plot)

Seaborn은 Pandas 데이터프레임을 기본적으로 지원합니다. 각 열을 series로 전닿하는 대신 열 이름을 제공하고 `data`인자를 사용하여 데이터 프레임을 지정할 수 있습니다.

````python
plt.title('Sepal Dimensions')
sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                hue='species',
                s=100,
                data=flowers_df);
````

## Histogram(히스토그램)

히스토그램은 값 범위를 따라 간격을 만들고 각 간견의 관측치 수를 나타내는 수직 막대를 표시하여 변수의 분포를 나타냅니다.  
예를 들어 flowers dataset에서 꽃받침 너비 값의 분포를 시각화 해 보겠습니다. `plt.hist` 함수를 사용하여 히스토그램을 생성 할 수 있습니다.

````python
# Load data into a Pandas dataframe
flowers_df = sns.load_dataset("iris")
````

````python
flowers_df.sepal_width
````

````python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
````

sepal 너비가 2.0~4.5 범위에 있고 약 35개의 값이 2.9~3.1 범위에 있다는 것을 한 눈에 알 수 있습니다.

### Controlling the size and number of bins (bin 크기 및 개수 제어)

bin 인자를 사용하여 그래프의 수와 크기를 조절할 수 있습니다.

````python
# bin 수 지정
plt.hist(flowers_df.sepal_width, bins=5);
````

````python
import numpy as np

# 각 bin의 경계 지정
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25));
````

````python
# 크기가 다른 bin
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4.5]); # group1 : 1~3, group2 : 3~4, group : 4~4.5
````

### Multiple Histograms(다중 히스토그램)

선형 차트와 마찬가지로 단일 차트에 여러 히스토그램을 그릴 수 있습니다.  
각 히스토그램의 막대가 다른 히스토그램을 숨기지 않도록 `alpha` 인자를 사용하여 각 히스토그램의 불투명도를 조절할 수 있습니다.

각각의 꽃 종에 대한 개별적인 히스토그램을 그려봅시다.

````python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
````

````python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
````

여러 개의 히스토그램을 서로 겹쳐 쌓을 수도 있습니다.

````python
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
````

## Bar Chart(막대형 차트)

막대형 차트는 꺽은선형 차트와 매우 유사합니다. 즉, 일련의 값을 표시합니다. 그러나 선으로 연결된 점이 아니라 각 값에 대해 막대가 표시됩니다. `plt.bar` 함수를 사용하여 막대 차트를 그릴 수 있습니다.

````python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
````

````python
plt.bar(years, oranges);
````

`plt.bar`의 `bottom` 인자를 사용하여 히스토그램처럼 서로 위로 막대를 쌓을 수 있습니다.

````python
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);
````

## Heatmap(히트맵)

히트맵은 색상을 이용하여 매트릭스 혹은 표와 같은 2차원 데이터를 시각화하는데 사용됩니다.  
`flights` 데이터셋을 사용하여 12년동안 공항에서 매달 발생하는 승객들의 수를 아래 예시에서 시각화할 것입니다.

````python
flights_df = sns.load_dataset("flights").pivot("month", "year", "passengers") #pivot은 여러 분류로 섞인 행 데이터를 열 데이터로 회전시킴
````

````python
flights_df
````

`flights_df`는 한 달에 한 행, 한 열이 있는 matrix로, 한 해 중 특정 달에 공항을 방문한 승객의 수를 나타냅니다.  
`sns.heatmap`함수를 사용하여 공황을 방문하는 승객의 수를 시각화 할 수 있습니다.

````python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df);
````

위 그래프의 밝은 색상은 더 많은 이용객이 방문했음을 의미합니다. 그래프를 통해 2 가지를 추론 할 수 있습니다.:

- 7~8월경에 공항 이용객이 높다.
- 각 달의 공항 이용객은 해마다 증가하는 경향이 있다.

각 블록에 `annot=True`를 설정하여 실제 값을 표시 할 수 있고 `cmap` 인자로 팔레트의 색을 바꿀 수 있습니다.

````python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df, fmt="d", annot=True);
````

````python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df, fmt="d", annot=True, cmap='Blues');
````

## Plotting multiple charts in a grid(gird에 여러 차트 표시)

Mattplotlib 와 Seaborn도 평면에 일련의 축을 반환하는 `plt.subplot`함수를 사용하여 격자 형식의 다중 그래프를 그릴 수 있습니다.

````python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))

# Use the axes for plotting
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].legend(['Apples', 'Oranges']);
axes[0,0].set_title('Crop Yields in Kanto')


# Pass the axes into seaborn
axes[0,1].set_title('Sepal Length vs. Sepal Width')
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species, 
                s=100, 
                ax=axes[0,1]);

# Use the axes for plotting
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);

# Pass the axes into seaborn
axes[1,0].set_title('Restaurant bills')
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, ax=axes[1,0]);

# Pass the axes into seaborn
axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1]);

# Plot an image using the axes
axes[1,2].set_title('Data Science Meme')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.tight_layout(pad=2);
````

해당 페이지의 대한 함수 목록은 해당 주소를 참조하기 바랍니다.: https://matplotlib.org/3.3.1/api/axes_api.html#the-axes-class .
